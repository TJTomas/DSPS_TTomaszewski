```python
import numpy as np
from matplotlib import pylab as plt
```

```python
np.random.seed(302)
x = np.random.randn(100) +12
```

```python
x.mean()
```

```python
x.std()
```

```python
plt.hist(x, bins = 10)
```

```python
x_minmax = x - x.min()
plt.hist(x_minmax, bins = 10)
```

```python
x_minmax = x_minmax / x_minmax.max()
plt.hist(x_minmax)
```

```python
x_standard = x - x.mean()
plt.hist(x_standard, bins=10)
```
